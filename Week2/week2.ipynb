```python
import matplotlib.pyplot as plt
x1=[]
y1=[]

for i in range(1,10):
    x1.append(i)
    y1.append(i+5)
x2=[]
y2=[]

for i in range(1,10):
    x2.append(i)
    y2.append(i*i)
plt.plot(x1,y1,marker="o",alpha=1,color="purple")
plt.plot(x2,y2,marker="o",alpha=1,color="black")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```
